```python
import os
import json
import pandas as pd
```

```python
annotations_df = pd.DataFrame()
```

```python
# annotation folder list
folder_list = [r"D:\Capstone_Data_Annotations\Annotation Stuff\MLFigs_Labeled_Oct_26_Chris",
               r"D:\Capstone_Data_Annotations\Annotation Stuff\MLFigsLabeled_Nov_03",
               r"D:\Capstone_Data_Annotations\Annotation Stuff\MLFigsLabeled_Nov_09_Chris",
               r"D:\Capstone_Data_Annotations\Annotation Stuff\MLFigsLabeled_Nov_09_Chris"]
```

```python
# ḷist of .json files residing in the folder
folder_json_data_list = []
file_name_list = []
```

```python
for folder_path in folder_list:
    df = pd.DataFrame()
    # List all files in the folder
    file_list = os.listdir(folder_path)
    # Iterate through the files
    for file_name in file_list:
        # Check if the file has a .json extension
        if file_name.endswith(".json"):
            file_path = os.path.join(folder_path, file_name)

            with open(file_path, 'r') as file:
                json_data = json.load(file)
            folder_json_data_list.append(json_data)
            file_name_list.append(file_name)
    df['file_name'] = file_name_list
    df['annotations'] = folder_json_data_list
    df['folder_name'] = [folder_path]*len(file_name_list)
    final_df = pd.concat([df.drop(['annotations'], axis=1), df['annotations'].apply(pd.Series)], axis=1)
            
    annotations_df = pd.concat([annotations_df, final_df])
            
```
